Сделайте кластеризацию данных методом `KMeans` (собственная имплементация и готовая реализация), обучив модель на данных со сниженной за счет PCA размерностью. В данном случае мы подскажем, что нужно искать именно 6 кластеров, но в общем случае мы не будем знать, сколько кластеров надо искать.

Параметры:

- **n_clusters** = n_classes (число уникальных меток целевого класса)
- **n_init** = 100
- **random_state** = RANDOM_STATE (для воспроизводимости результата)

Остальные параметры со значениями по умолчанию.

````python
class KMeansCustom:
    def __init__(self, n_clusters, n_init=100, max_iter=300, random_state=None):
        self.n_clusters = n_clusters
        self.n_init = n_init
        self.max_iter = max_iter
        self.random_state = random_state
        self.labels_ = None
        self.cluster_centers_ = None
        self.inertia_ = None

    def fit(self, X):
        np.random.seed(self.random_state)
        best_inertia = np.inf
        best_centers = None
        best_labels = None

        for init in range(self.n_init):
            random_indices = np.random.choice(X.shape[0], self.n_clusters, replace=False)
            centers = X[random_indices].copy()

            for iteration in range(self.max_iter):
                distances = np.sqrt(((X - centers[:, np.newaxis])**2).sum(axis=2))
                labels = np.argmin(distances, axis=0)

                new_centers = np.array([X[labels == k].mean(axis=0)
                                       for k in range(self.n_clusters)])

                if np.allclose(centers, new_centers):
                    break

                centers = new_centers

            inertia = sum([np.sum((X[labels == k] - centers[k])**2)
                          for k in range(self.n_clusters)])

            if inertia < best_inertia:
                best_inertia = inertia
                best_centers = centers
                best_labels = labels

        self.cluster_centers_ = best_centers
        self.labels_ = best_labels
        self.inertia_ = best_inertia

        return self

    def predict(self, X):
        distances = np.sqrt(((X - self.cluster_centers_[:, np.newaxis])**2).sum(axis=2))
        return np.argmin(distances, axis=0)
````

````python
print("Обучение собственной реализации KMeans...")
kmeans_custom = KMeansCustom(n_clusters=n_classes, n_init=100, random_state=RANDOM_STATE)
kmeans_custom.fit(X_pca)

print(f"Собственная реализация - Inertia: {kmeans_custom.inertia_:.2f}")

print("\nОбучение sklearn KMeans...")
kmeans_sklearn = KMeans(n_clusters=n_classes, n_init=100, random_state=RANDOM_STATE)
kmeans_sklearn.fit(X_pca)

print(f"Sklearn реализация - Inertia: {kmeans_sklearn.inertia_:.2f}")

print(f"\nРазница в Inertia: {abs(kmeans_custom.inertia_ - kmeans_sklearn.inertia_):.2f}")
````

Попробуйте использовать DBSCAN в качестве алгоритма кластеризации и метод понижения размерности tSNE.

````python
from sklearn.cluster import DBSCAN
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, random_state=RANDOM_STATE, perplexity=30)
X_tsne = tsne.fit_transform(X_scaled)

dbscan = DBSCAN(eps=5.0, min_samples=10)
cluster_labels_dbscan = dbscan.fit_predict(X_tsne)

n_clusters_dbscan = len(set(cluster_labels_dbscan)) - (1 if -1 in cluster_labels_dbscan else 0)
n_noise = list(cluster_labels_dbscan).count(-1)

print(f"\nРезультаты DBSCAN:")
print(f"  - Найдено кластеров: {n_clusters_dbscan}")
print(f"  - Точек-выбросов (шум): {n_noise}")
print(f"  - eps: {dbscan.eps}")
print(f"  - min_samples: {dbscan.min_samples}")

plt.figure(figsize=(12, 5))

# Подзаголовок 1: Истинные метки
plt.subplot(1, 2, 1)
scatter1 = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, s=20, cmap='viridis', alpha=0.6)
plt.colorbar(scatter1)
plt.title('t-SNE: Истинные метки классов')
plt.xlabel('t-SNE компонента 1')
plt.ylabel('t-SNE компонента 2')

# Подзаголовок 2: DBSCAN кластеры
plt.subplot(1, 2, 2)
scatter2 = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=cluster_labels_dbscan, s=20, cmap='viridis', alpha=0.6)
plt.colorbar(scatter2)
plt.title(f'DBSCAN кластеризация (кластеров: {n_clusters_dbscan}, шум: {n_noise})')
plt.xlabel('t-SNE компонента 1')
plt.ylabel('t-SNE компонента 2')

plt.tight_layout()
plt.show()
print("\n4. Оценка качества кластеризации:")

mask_no_noise = cluster_labels_dbscan != -1

if mask_no_noise.sum() > 0:
    ari_dbscan = metrics.adjusted_rand_score(y[mask_no_noise],
                                              cluster_labels_dbscan[mask_no_noise])
    print(f"  - ARI (без учета шума): {ari_dbscan:.4f}")
else:
    print("  - Все точки классифицированы как шум!")

ari_dbscan_full = metrics.adjusted_rand_score(y, cluster_labels_dbscan)
print(f"  - ARI (с учетом шума): {ari_dbscan_full:.4f}")

print("РЕЗУЛЬТАТЫ:")
print(f"t-SNE успешно снизил размерность до 2D")
print(f"DBSCAN нашел {n_clusters_dbscan} кластеров")
print(f"Выявлено {n_noise} точек-выбросов")
````
